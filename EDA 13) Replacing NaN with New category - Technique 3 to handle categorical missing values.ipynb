# Technique 3) Replacing NaN with New category
#### if frequent categories are multiple, then NaN is replaced with New category

```python
import pandas as pd
df = pd.read_csv('House Prices_train.csv')
# Download Link: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=train.csv
df
```

```python
# lets focus on 4 particular columns / features
df = pd.read_csv('House Prices_train.csv',usecols=['BsmtQual', 'FireplaceQu', 'GarageType', 'SalePrice'])
df
# 'BsmtQual', 'FireplaceQu', 'GarageType' are Categorical features
# 'SalePrice' is dependent feature
```

```python
import numpy as np
# creating a new function
def impute_nan(df,variable):
    df[variable+'_New_Category'] = np.where(df[variable].isnull(),'Missing',df[variable])
#"missing value" is a new category
```

```python
# For loop , as explained in last video
for feature in ['BsmtQual', 'FireplaceQu', 'GarageType']:
    impute_nan(df,feature)
df.head(25)
```

```python
#As New features are generated by replacing Nan values,
# So, Old features are not needed. Lets DROP them

df = df.drop(['BsmtQual', 'FireplaceQu', 'GarageType'],axis=1) # "axis=1" mentiones the axis to find the features
df.head(25)
```
